```sql
USE CATALOG EDW_WH;

-- ============================================================
-- 1. stg_accounts
-- ============================================================
COPY INTO stgdb.stg_accounts
FROM (
    SELECT
        AccountID::VARCHAR(50) AS AccountID,
        AccountType::VARCHAR(50) AS AccountType,
        Balance::DECIMAL(18,2) AS Balance,
        CreditScore::INT AS CreditScore,
        Currency::VARCHAR(10) AS Currency,
        CustomerID::VARCHAR(50) AS CustomerID,
        DateOpened::DATE AS DateOpened,
        ManagerID::VARCHAR(50) AS ManagerID,
        ODLimit::DECIMAL(18,2) AS ODLimit
    FROM '/Volumes/edw_wh/stgdb/raw_data/accounts.csv'
)
FILEFORMAT = CSV
FORMAT_OPTIONS (
    'header'      = 'true',
    'delimiter'   = ','
);

select * from stgdb.stg_accounts;

select * from stgdb.stg_branches;

copy into stgdb.stg_branches
from (
    select 
        Address,
        BranchID,
        BranchName,
        City,
        State,
        Zipcode
        from '/Volumes/edw_wh/stgdb/raw_data/branches.csv'
)
FILEFORMAT = CSV
FORMAT_OPTIONS (
    'header'      = 'true',
    'delimiter'   = ','
);

desc stgdb.stg_creditcard;

copy into stgdb.stg_creditcard
from (
    select 
        Balance::decimal(18,2) as Balance,
        BillCycle,
        CardID,
        CardNumber,
        CardType,
        CreditLimit::decimal(18,2) as CreditLimit,
        CustomerID,
        ExpirationDate::date as ExpirationDate,
        InterestRate::decimal(5,2) as InterestRate,
        IssueDate::date as IssueDate,
        Status
        from '/Volumes/edw_wh/stgdb/raw_data/creditcard.csv'
)
FILEFORMAT = CSV
FORMAT_OPTIONS (
    'header'      = 'true',
    'delimiter'   = ','
);

select * from  stgdb.stg_cust_profile;

truncate table stgdb.stg_cust_profile;

copy into stgdb.stg_cust_profile
from (
    select 
        Address::string as Address,
        BranchID,
        CustomerID,
        DateOfBirth::date as DateOfBirth,
        Email,
        FirstName,
        LastName,
        CASE
    WHEN PhoneNumber LIKE '001-%'
        THEN REGEXP_REPLACE(PhoneNumber, '^001', '1')
    WHEN LENGTH(PhoneNumber) > 20
        THEN LEFT(PhoneNumber, 20)
    ELSE PhoneNumber
END AS PhoneNumber
        from '/Volumes/edw_wh/stgdb/raw_data/cust.csv'
)
FILEFORMAT = CSV
FORMAT_OPTIONS (
    'header'      = 'true',
    'delimiter'   = ','
)
COPY_OPTIONS (
    'mergeSchema' = 'true',
    'force'       = 'true'    -- reloads file even if previously processed
);

select * from stgdb.stg_employees;

copy into stgdb.stg_employees
from (
    select 
        BranchID,
        EmployeeID,
        FirstName,
        Hiredate::date as Hiredate,
        LastName,
        ManagerID,
        Position
        from '/Volumes/edw_wh/stgdb/raw_data/employee.csv'
)
FILEFORMAT = CSV
FORMAT_OPTIONS (
    'header'      = 'true',
    'delimiter'   = ','
);

select * from stgdb.stg_loans;

desc stgdb.stg_loans;;


copy into stgdb.stg_loans
from (
    select 
        Amount::decimal(18,2) as Amount,
        Collateral,
        CustomerID,
        EndDate::date as EndDate,
        InterestRate::decimal(5,2) as InterestRate,
        LoanID,
        loanType,
        PaymentFrequency,
        StartDate::date as StartDate,
        Status
        from '/Volumes/edw_wh/stgdb/raw_data/loans.csv'
)
FILEFORMAT = CSV
FORMAT_OPTIONS (
    'header'      = 'true',
    'delimiter'   = ','
);


select * from edw_wh.stgdb.stg_payments

desc edw_wh.stgdb.stg_payments


copy into edw_wh.stgdb.stg_payments
from (
    select 
        Amount::decimal(18,2) as Amount,
        AuditTrial,
        ClearingSystem,
        Currency,
        CustomerSegment,
        Description,
        ExchangeRate::decimal(10,4) as ExchangeRate,
        Fee::decimal(18,2) as Fee,
        FromAccountID,
        MerchantName,
        PaymentDate::date as PaymentDate,
        PaymentID,
        PaymentType,
        ToAccountID
        from '/Volumes/edw_wh/stgdb/raw_data/payments.csv'
)
FILEFORMAT = CSV
FORMAT_OPTIONS (
    'header'      = 'true',
    'delimiter'   = ','
);


desc stgdb.stg_transactions;

copy into edw_wh.stgdb.stg_transactions
from (
    select 
        AccountID,
        Amount::decimal(18,2) as Amount,
        Currency,
        Description,
        EventTs::timestamp as EventTs,
        Status,
        Suspecious::boolean as Suspecious,
        TransactionDate::date as TransactionDate,
        TransactionFee::decimal(18,2) as TransactionFee,
        TransactionID,
        TransactionType
        from '/Volumes/edw_wh/stgdb/raw_data/transactions.csv'
)
FILEFORMAT = CSV
FORMAT_OPTIONS (
    'header'      = 'true',
    'delimiter'   = ','
)
COPY_OPTIONS (
    'mergeSchema' = 'true');

select * from edw_wh.stgdb.stg_transactions
```
