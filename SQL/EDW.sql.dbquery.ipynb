```sql
-- ============================================================
-- CATALOG
-- ============================================================
USE CATALOG EDW_WH;

-- ============================================================
-- DROP DATABASES
-- ============================================================
DROP DATABASE IF EXISTS stgdb   CASCADE;
DROP DATABASE IF EXISTS odsdb   CASCADE;
DROP DATABASE IF EXISTS edwdb   CASCADE;
DROP DATABASE IF EXISTS trans_mart   CASCADE;
DROP DATABASE IF EXISTS payment_mart CASCADE;
DROP DATABASE IF EXISTS cc_mart      CASCADE;

-- ============================================================
-- STGDB - STAGING LAYER
-- ============================================================
CREATE DATABASE stgdb;

CREATE TABLE stgdb.stg_accounts (
    AccountID       VARCHAR(50),
    AccountType     VARCHAR(50),
    Balance         DECIMAL(18,2),
    CreditScore     INT,
    Currency        VARCHAR(10),
    CustomerID      VARCHAR(50),
    DateOpened      DATE,
    ManagerID       VARCHAR(50),
    ODLimit         DECIMAL(18,2)
) USING DELTA;

CREATE TABLE stgdb.stg_transactions (
    AccountID       VARCHAR(50),
    Amount          DECIMAL(18,2),
    Currency        VARCHAR(10),
    Description     STRING,
    EventTs         TIMESTAMP,
    Status          VARCHAR(50),
    Suspicious      BOOLEAN,
    TransactionDate DATE,
    TransactionFee  DECIMAL(18,2),
    TransactionID   VARCHAR(50),
    TransactionType VARCHAR(50)
) USING DELTA;

CREATE TABLE stgdb.stg_payments (
    Amount          DECIMAL(18,2),
    AuditTrial      STRING,
    ClearingSystem  VARCHAR(50),
    Currency        VARCHAR(10),
    CustomerSegment VARCHAR(50),
    Description     STRING,
    ExchangeRate    DECIMAL(10,4),
    Fee             DECIMAL(18,2),
    FromAccountID   VARCHAR(50),
    MerchantName    VARCHAR(100),
    PaymentDate     DATE,
    PaymentID       VARCHAR(50),
    PaymentType     VARCHAR(50),
    ToAccountID     VARCHAR(50)
) USING DELTA;

CREATE TABLE stgdb.stg_creditcard (
    Balance         DECIMAL(18,2),
    BillCycle       VARCHAR(20),
    CardID          VARCHAR(50),
    CardNumber      VARCHAR(50),
    CardType        VARCHAR(50),
    CreditLimit     DECIMAL(18,2),
    CustomerID      VARCHAR(50),
    ExpirationDate  DATE,
    InterestRate    DECIMAL(5,2),
    IssueDate       DATE,
    Status          VARCHAR(20)
) USING DELTA;

CREATE TABLE stgdb.stg_loans (
    Amount          DECIMAL(18,2),
    Collateral      VARCHAR(100),
    CustomerID      VARCHAR(50),
    EndDate         DATE,
    InterestRate    DECIMAL(5,2),
    LoanID          VARCHAR(50),
    LoanType        VARCHAR(50),
    PaymentFrequency VARCHAR(20),
    StartDate       DATE,
    Status          VARCHAR(20)
) USING DELTA;

CREATE TABLE stgdb.stg_cust_profile (
    Address         STRING,
    BranchID        VARCHAR(50),
    CustomerID      VARCHAR(50),
    DateOfBirth     DATE,
    Email           VARCHAR(100),
    FirstName       VARCHAR(100),
    LastName        VARCHAR(100),
    PhoneNumber     VARCHAR(20)
) USING DELTA;

CREATE TABLE stgdb.stg_branches (
    Address         STRING,
    BranchID        VARCHAR(50),
    BranchName      VARCHAR(100),
    City            VARCHAR(100),
    State           VARCHAR(100),
    Zipcode         VARCHAR(20)
) USING DELTA;

CREATE TABLE stgdb.stg_employees (
    BranchID        VARCHAR(50),
    EmployeeID      VARCHAR(50),
    FirstName       VARCHAR(100),
    Hiredate        DATE,
    LastName        VARCHAR(100),
    ManagerID       VARCHAR(50),
    Position        VARCHAR(100)
) USING DELTA;

-- ============================================================
-- ODSDB - OPERATIONAL DATA STORE LAYER
-- ============================================================
CREATE DATABASE odsdb;

CREATE TABLE odsdb.ods_accounts (
    AccountID       VARCHAR(50),
    AccountType     VARCHAR(50),
    Balance         DECIMAL(18,2),
    CreditScore     INT,
    Currency        VARCHAR(10),
    CustomerID      VARCHAR(50),
    DateOpened      DATE,
    ManagerID       VARCHAR(50),
    ODLimit         DECIMAL(18,2),
    load_dt         DATE,
    load_ts         TIMESTAMP
) USING DELTA;

CREATE TABLE odsdb.ods_transactions (
    AccountID       VARCHAR(50),
    Amount          DECIMAL(18,2),
    Currency        VARCHAR(10),
    Description     STRING,
    EventTs         TIMESTAMP,
    Status          VARCHAR(50),
    Suspicious      BOOLEAN,
    TransactionDate DATE,
    TransactionFee  DECIMAL(18,2),
    TransactionID   VARCHAR(50),
    TransactionType VARCHAR(50),
    load_dt         DATE,
    load_ts         TIMESTAMP
) USING DELTA;

CREATE TABLE odsdb.ods_payments (
    Amount          DECIMAL(18,2),
    AuditTrial      STRING,
    ClearingSystem  VARCHAR(50),
    Currency        VARCHAR(10),
    CustomerSegment VARCHAR(50),
    Description     STRING,
    ExchangeRate    DECIMAL(10,4),
    Fee             DECIMAL(18,2),
    FromAccountID   VARCHAR(50),
    MerchantName    VARCHAR(100),
    PaymentDate     DATE,
    PaymentID       VARCHAR(50),
    PaymentType     VARCHAR(50),
    ToAccountID     VARCHAR(50),
    load_dt         DATE,
    load_ts         TIMESTAMP
) USING DELTA;

CREATE TABLE odsdb.ods_creditcard (
    Balance         DECIMAL(18,2),
    BillCycle       VARCHAR(20),
    CardID          VARCHAR(50),
    CardNumber      VARCHAR(50),
    CardType        VARCHAR(50),
    CreditLimit     DECIMAL(18,2),
    CustomerID      VARCHAR(50),
    ExpirationDate  DATE,
    InterestRate    DECIMAL(5,2),
    IssueDate       DATE,
    Status          VARCHAR(20),
    load_dt         DATE,
    load_ts         TIMESTAMP
) USING DELTA;

CREATE TABLE odsdb.ods_loans (
    Amount          DECIMAL(18,2),
    Collateral      VARCHAR(100),
    CustomerID      VARCHAR(50),
    EndDate         DATE,
    InterestRate    DECIMAL(5,2),
    LoanID          VARCHAR(50),
    LoanType        VARCHAR(50),
    PaymentFrequency VARCHAR(20),
    StartDate       DATE,
    Status          VARCHAR(20),
    load_dt         DATE,
    load_ts         TIMESTAMP
) USING DELTA;

CREATE TABLE odsdb.ods_cust_profile (
    Address         STRING,
    BranchID        VARCHAR(50),
    CustomerID      VARCHAR(50),
    DateOfBirth     DATE,
    Email           VARCHAR(100),
    FirstName       VARCHAR(100),
    LastName        VARCHAR(100),
    PhoneNumber     VARCHAR(20),
    load_dt         DATE,
    load_ts         TIMESTAMP
) USING DELTA;

CREATE TABLE odsdb.ods_branches (
    Address         STRING,
    BranchID        VARCHAR(50),
    BranchName      VARCHAR(100),
    City            VARCHAR(100),
    State           VARCHAR(100),
    Zipcode         VARCHAR(20),
    load_dt         DATE,
    load_ts         TIMESTAMP
) USING DELTA;

CREATE TABLE odsdb.ods_employees (
    BranchID        VARCHAR(50),
    EmployeeID      VARCHAR(50),
    FirstName       VARCHAR(100),
    Hiredate        DATE,
    LastName        VARCHAR(100),
    ManagerID       VARCHAR(50),
    Position        VARCHAR(100),
    load_dt         DATE,
    load_ts         TIMESTAMP
) USING DELTA;

-- ============================================================
-- EDWDB - EDW LAYER
-- ============================================================
CREATE DATABASE edwdb;

CREATE TABLE edwdb.dim_customers (
    CustomerID      VARCHAR(50),
    FirstName       VARCHAR(100),
    LastName        VARCHAR(100),
    Email           VARCHAR(100),
    PhoneNumber     VARCHAR(20),
    Address         STRING,
    DateOfBirth     DATE,
    BranchID        VARCHAR(50),
    load_dt         DATE,
    load_ts         TIMESTAMP,
    effective_date  DATE
) USING DELTA;

CREATE TABLE edwdb.dim_branches (
    Address         STRING,
    BranchID        VARCHAR(50),
    BranchName      VARCHAR(100),
    City            VARCHAR(100),
    State           VARCHAR(100),
    Zipcode         VARCHAR(20),
    load_dt         DATE,
    load_ts         TIMESTAMP,
    start_date      DATE,
    end_date        DATE,
    is_current      TINYINT
) USING DELTA;

CREATE TABLE edwdb.dim_employees (
    BranchID        VARCHAR(50),
    EmployeeID      VARCHAR(50),
    FirstName       VARCHAR(100),
    Hiredate        DATE,
    LastName        VARCHAR(100),
    ManagerID       VARCHAR(50),
    Position        VARCHAR(100),
    load_dt         DATE,
    load_ts         TIMESTAMP
) USING DELTA;

CREATE TABLE edwdb.dim_loans (
    Amount          DECIMAL(18,2),
    Collateral      VARCHAR(100),
    CustomerID      VARCHAR(50),
    EndDate         DATE,
    InterestRate    DECIMAL(5,2),
    LoanID          VARCHAR(50),
    LoanType        VARCHAR(50),
    PaymentFrequency VARCHAR(20),
    StartDate       DATE,
    Status          VARCHAR(20),
    load_dt         DATE,
    load_ts         TIMESTAMP
) USING DELTA;

CREATE TABLE edwdb.fact_loans (
    fact_loan_id        BIGINT        GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
    LoanID              BIGINT        NOT NULL,
    CustomerID          INT           NOT NULL,
    BranchID            INT           NOT NULL,
    Amount              DECIMAL(18,2) NOT NULL,
    InterestRate        DECIMAL(5,2)  NOT NULL,
    StartDate           DATE          NOT NULL,
    EndDate             DATE          NOT NULL,
    PaymentFrequency    VARCHAR(20),
    Status              VARCHAR(20),
    OutstandingBalance  DECIMAL(18,2),
    LoanDurationMonths  INT,
    RiskIndicator       VARCHAR(10),
    HighValueFlag       BOOLEAN,
    load_dt             DATE          NOT NULL,
    load_ts             TIMESTAMP
)
USING DELTA
PARTITIONED BY (load_dt)
TBLPROPERTIES (
    'delta.feature.allowColumnDefaults' = 'supported',
    'delta.autoOptimize.optimizeWrite'  = 'true',
    'delta.autoOptimize.autoCompact'    = 'true'
)
COMMENT 'EDW Fact table for loan transactions';

ALTER TABLE edwdb.fact_loans
ALTER COLUMN load_ts SET DEFAULT CURRENT_TIMESTAMP();

-- ============================================================
-- TRANS_MART - TRANSACTION MART
-- ============================================================
CREATE DATABASE trans_mart;

CREATE TABLE trans_mart.fact_transactions (
    AccountID       VARCHAR(50),
    Amount          DECIMAL(18,2),
    Currency        VARCHAR(10),
    Description     STRING,
    EventTs         TIMESTAMP,
    Status          VARCHAR(50),
    Suspicious      BOOLEAN,
    TransactionDate DATE,
    TransactionFee  DECIMAL(18,2),
    TransactionID   VARCHAR(50),
    TransactionType VARCHAR(50),
    load_dt         DATE,
    load_ts         TIMESTAMP,
    transaction_flag VARCHAR(10)
) USING DELTA;

-- ============================================================
-- PAYMENT_MART - PAYMENT MART
-- ============================================================
CREATE DATABASE payment_mart;

CREATE TABLE payment_mart.fact_payments (
    Amount              DECIMAL(18,2),
    AuditTrial          STRING,
    ClearingSystem      VARCHAR(50),
    Currency            VARCHAR(10),
    CustomerSegment     VARCHAR(50),
    Description         STRING,
    ExchangeRate        DECIMAL(10,4),
    Fee                 DECIMAL(18,2),
    FromAccountID       VARCHAR(50),
    MerchantName        VARCHAR(100),
    PaymentDate         DATE,
    PaymentID           VARCHAR(50),
    PaymentType         VARCHAR(50),
    ToAccountID         VARCHAR(50),
    load_dt             DATE,
    load_ts             TIMESTAMP,
    AmountInBaseCurrency DECIMAL(18,2)
) USING DELTA;

-- ============================================================
-- CC_MART - CREDIT CARD MART
-- ============================================================
CREATE DATABASE cc_mart;

CREATE TABLE cc_mart.fact_creditcard (
    fact_creditcard_id  BIGINT        GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
    customerid          INT,
    loanid              BIGINT,
    employeeid          INT,
    firstname           VARCHAR(100),
    phonenumber         VARCHAR(20),
    cardid              BIGINT,
    cardtype            VARCHAR(50),
    balance             DECIMAL(15,2),
    creditlimit         DECIMAL(15,2),
    billcycle           VARCHAR(20),
    issuedate           DATE,
    utilization_percent DECIMAL(5,2),
    load_dt             DATE,
    load_ts             TIMESTAMP
)
USING DELTA
TBLPROPERTIES (
    'delta.feature.allowColumnDefaults' = 'supported',
    'delta.autoOptimize.optimizeWrite'  = 'true',
    'delta.autoOptimize.autoCompact'    = 'true'
)
COMMENT 'Credit Card Mart fact table';

ALTER TABLE cc_mart.fact_creditcard
ALTER COLUMN load_ts SET DEFAULT CURRENT_TIMESTAMP();
```
